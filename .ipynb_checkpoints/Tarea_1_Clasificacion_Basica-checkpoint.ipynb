## Parte 1: Importar librerías y cargar datos

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron, LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from tensorflow.keras.datasets import mnist
import seaborn as sns

# Configuración para reproducibilidad
np.random.seed(42)
```

```python
import os
import shutil

# Limpiar caché de scikit-learn
cache_path = os.path.expanduser('~\\scikit_learn_data')
print(f"Ruta de caché: {cache_path}")

if os.path.exists(cache_path):
    shutil.rmtree(cache_path)
    print("✓ Caché eliminada")
else:
    print("No hay caché")
```

## Parte 2: Cargar y preparar el dataset MNIST

```python
# Cargar MNIST
print("Cargando MNIST...")
# mnist = fetch_openml(data_id=554, parser='auto')
# X, y = mnist["data"].values, mnist["target"].values.astype(int)
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Unir train + test para imitar fetch_openml
X = np.concatenate([X_train, X_test])
y = np.concatenate([y_train, y_test]).astype(int)

# Aplanar imágenes (28x28 -> 784)
X = X.reshape(-1, 784)

# TODO: Filtrar solo los dígitos 0 y 1
# Pista: Usa np.isin() para crear una máscara booleana
mask =  np.isin(y, [0, 1])
X_binary = X[mask]
y_binary = y[mask]

print(f"Datos filtrados: {X_binary.shape[0]} ejemplos")
print(f"Distribución de clases: {np.bincount(y_binary)}")
```

## Parte 3: Visualizar algunos ejemplos

```python
# TODO: Visualizar 10 ejemplos aleatorios del dataset
# Pista: Usa plt.subplot() y reshape() para mostrar las imágenes 28x28

fig, axes = plt.subplots(2, 5, figsize=(12, 5))
indices = np.random.choice(len(X_binary), 10, replace=False)

for i, ax in enumerate(axes.flat):
    # COMPLETA AQUÍ: Mostrar la imagen correspondiente
    ax.imshow(X_binary[indices[i]].reshape(28, 28), cmap='gray')
    ax.set_title(f"Etiqueta: {y_binary[indices[i]]}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```

## Parte 4: Dividir datos y normalizar

```python
# TODO: Dividir en train (70%), validation (15%) y test (15%)
# Pista: Usa train_test_split dos veces

X_temp, X_test, y_temp, y_test = train_test_split(
    X_binary, y_binary, 
    test_size=0.15, 
    random_state=42, 
    stratify=y_binary
)

X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, 
    test_size=0.17647, 
    random_state=42, 
    stratify=y_temp
)

print(f"Train: {X_train.shape[0]}, Validation: {X_val.shape[0]}, Test: {X_test.shape[0]}")
```

```python
# TODO: Normalizar los datos usando StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)
```

## Parte 5: Entrenar Perceptrón

```python
# TODO: Crear y entrenar un modelo Perceptron
# Pista: max_iter=1000, random_state=42

perceptron = Perceptron(
    max_iter=1000,       # Número máximo de iteraciones
    random_state=42,     # Para reproducibilidad
    tol=1e-3             # Tolerancia para criterio de parada
)
perceptron.fit(X_train_scaled, y_train)

# Evaluar
y_pred_train = perceptron.predict(X_train_scaled)
y_pred_val = perceptron.predict(X_val_scaled)

print(f"Accuracy en Train: {accuracy_score(y_train, y_pred_train):.4f}")
print(f"Accuracy en Validation: {accuracy_score(y_val, y_pred_val):.4f}")
```

## Parte 6: Entrenar Regresión Logística

```python
# TODO: Crear y entrenar un modelo de Regresión Logística
# Pista: max_iter=1000, random_state=42

log_reg = LogisticRegression(
    max_iter=1000,      # Número máximo de iteraciones
    random_state=42,    # Para reproducibilidad
    solver='lbfgs'      # Solver recomendado para pequeños/múltiples datasets
)
log_reg.fit(X_train_scaled, y_train)

# Evaluar
y_pred_train_lr = log_reg.predict(X_train_scaled)
y_pred_val_lr = log_reg.predict(X_val_scaled)

print(f"Accuracy en Train: {accuracy_score(y_train, y_pred_train_lr):.4f}")
print(f"Accuracy en Validation: {accuracy_score(y_val, y_pred_val_lr):.4f}")
```

## Parte 7: Comparar modelos en Test

```python
# TODO: Evaluar ambos modelos en el conjunto de test
y_pred_test_p = perceptron.predict(X_test_scaled)
y_pred_test_lr = log_reg.predict(X_test_scaled)

print("="*50)
print("RESULTADOS EN TEST")
print("="*50)
print(f"\nPerceptrón - Accuracy: {accuracy_score(y_test, y_pred_test_p):.4f}")
print(f"Regresión Logística - Accuracy: {accuracy_score(y_test, y_pred_test_lr):.4f}")
```

## Parte 8: Matriz de Confusión

```python
# TODO: Crear y visualizar las matrices de confusión para ambos modelos

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Perceptrón
cm_p = confusion_matrix(y_test, y_pred_test_p)
sns.heatmap(cm_p, annot=True, fmt='d', ax=axes[0], cmap='Blues')
axes[0].set_title('Matriz de Confusión - Perceptrón')
axes[0].set_ylabel('Etiqueta Real')
axes[0].set_xlabel('Etiqueta Predicha')

# TODO: Completa para Regresión Logística
cm_lr = confusion_matrix(y_test, y_pred_test_lr)
sns.heatmap(cm_lr, annot=True, fmt='d', ax=axes[1], cmap='Greens')
axes[1].set_title('Matriz de Confusión - Regresión Logística')
axes[1].set_ylabel('Etiqueta Real')
axes[1].set_xlabel('Etiqueta Predicha')

plt.tight_layout()
plt.show()
```
