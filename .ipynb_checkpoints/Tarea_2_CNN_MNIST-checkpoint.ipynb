## Parte 1: Importar librerías y configuración

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
from tqdm import tqdm

# Configuración
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Usando dispositivo: {DEVICE}")

# Semillas para reproducibilidad
torch.manual_seed(42)
np.random.seed(42)

# Hiperparámetros
BATCH_SIZE = 64
LEARNING_RATE = 0.001
EPOCHS = 15
NUM_CLASSES = 10
```

## Parte 2: Cargar y preparar datos

```python
# Cargar MNIST
print("Cargando MNIST...")
mnist = fetch_openml('mnist_784', version=1, parser='auto')
X, y = mnist["data"].values, mnist["target"].values.astype(int)

# Usar subconjunto para acelerar
n_samples = 20000
indices = np.random.choice(len(X), n_samples, replace=False)
X = X[indices]
y = y[indices]

print(f"Dataset: {X.shape}")
print(f"Distribución de clases:\n{np.bincount(y)}")
```

## Parte 3: Visualizar ejemplos

```python
# TODO: Visualizar 10 ejemplos aleatorios (uno de cada dígito)
fig, axes = plt.subplots(2, 5, figsize=(15, 6))

for digit in range(10):
    # COMPLETA: Encuentra el primer índice donde y == digit
    idx = np.where(y == digit)[0][0]
    
    ax = axes[digit // 5, digit % 5]
    # COMPLETA: Muestra la imagen (reshape a 28x28)
    ax.imshow(X[idx].reshape(28, 28), cmap='gray')
    ax.set_title(f"Dígito: {digit}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```

## Parte 4: Preparar datos para PyTorch

```python
# TODO: Dividir en train (70%), validation (15%) y test (15%)
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, 
    test_size=0.15, 
    random_state=42, 
    stratify=y
)

X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, 
    test_size=0.15 / 0.85,  # ≈ 0.17647
    random_state=42, 
    stratify=y_temp
)

print(f"Train: {X_train.shape[0]}, Val: {X_val.shape[0]}, Test: {X_test.shape[0]}")
```

```python
# TODO: Normalizar y reshape para CNN
# Las CNNs esperan imágenes en formato (batch, channels, height, width)
# MNIST: (batch, 1, 28, 28)

# Normalizar a rango [0, 1]
X_train = X_train / 255.0
X_val = X_val / 255.0
X_test = X_test / 255.0

# TODO: Reshape de (N, 784) a (N, 1, 28, 28)
X_train = X_train.reshape(-1, 1, 28, 28)
X_val = X_val.reshape(-1, 1, 28, 28)
X_test = X_test.reshape(-1, 1, 28, 28)

print(f"Shape después de reshape: {X_train.shape}")
```

```python
# TODO: Crear DataLoaders de PyTorch
# Pista: Convierte a tensores FloatTensor para X y LongTensor para y

train_dataset = TensorDataset(
    torch.FloatTensor(X_train),
    torch.LongTensor(y_train)
)

val_dataset = TensorDataset(
    torch.FloatTensor(X_val),
    torch.LongTensor(y_val)
)

test_dataset = TensorDataset(
    torch.FloatTensor(X_test),
    torch.LongTensor(y_test)
)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

print(f"Número de batches - Train: {len(train_loader)}, Val: {len(val_loader)}, Test: {len(test_loader)}")
```

## Parte 5: Definir la arquitectura de la CNN

```python
# TODO: Implementar una CNN para clasificación de MNIST
# Arquitectura sugerida:
# - Conv1: 1 input channel, 32 output channels, kernel 3x3
# - ReLU + MaxPool 2x2
# - Conv2: 32 input channels, 64 output channels, kernel 3x3
# - ReLU + MaxPool 2x2
# - Flatten
# - FC1: 64*5*5 -> 128 neuronas + ReLU
# - Dropout(0.5)
# - FC2: 128 -> 10 (clases)

class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        
        # TODO: Define las capas convolucionales
        self.conv1 = nn.Conv2d(
            in_channels=1,
            out_channels=32,
            kernel_size=3,
            padding=1  # Para mantener dimensiones
        )
        
        self.conv2 = nn.Conv2d(
            in_channels=32,
            out_channels=64,
            kernel_size=3,
            padding=1
        )
        
        # Pooling layer
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        
        # TODO: Define las capas fully connected
        # Después de dos max pooling de 2x2, una imagen 28x28 se convierte en 7x7
        # Con 64 channels: 64 * 7 * 7 = 3136
        self.fc1 = nn.Linear(64 * 7 * 7, 128)
        self.fc2 = nn.Linear(128, NUM_CLASSES)
        
        # Dropout
        self.dropout = nn.Dropout(0.5)
    
    def forward(self, x):
        # TODO: Implementa el forward pass
        # Pista: usa F.relu() para activaciones
        
        # Bloque convolucional 1
        x = self.pool(F.relu(self.conv1(x)))
        
        # Bloque convolucional 2
        x = x = self.pool(F.relu(self.conv2(x))) # aplica conv2, relu y pool
        
        # Flatten
        x = x.view(x.size(0), -1)  # (batch_size, 64*7*7)
        
        # Capas fully connected
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x) # fc2 (sin activación, se usa CrossEntropyLoss)
        
        return x

# Crear el modelo
model = CNN().to(DEVICE)
print(model)

# Contar parámetros
total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"\nTotal de parámetros: {total_params:,}")
print(f"Parámetros entrenables: {trainable_params:,}")
```

## Parte 6: Configurar optimizador y función de pérdida

```python
# TODO: Configurar el optimizador (Adam) y la función de pérdida (CrossEntropyLoss)
optimizer = optim.Adam(
    model.parameters(),
    lr=0.001
)

criterion = nn.CrossEntropyLoss() # función de pérdida para clasificación multiclase
```

## Parte 7: Función de entrenamiento

```python
# TODO: Implementar la función de entrenamiento
def train_epoch(model, train_loader, criterion, optimizer, device):
    """
    Entrena el modelo por una época
    """
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    
    for images, labels in tqdm(train_loader, desc="Entrenando"):
        # TODO: Mover datos al dispositivo
        images = images.to(device)
        labels = labels.to(device)
        
        # TODO: Forward pass
        optimizer.zero_grad()
        outputs = model(images) # pasar images por el modelo
        loss = criterion(outputs, labels) # calcular pérdida
        
        # TODO: Backward pass y optimización
        loss.backward()
        optimizer.step()
        
        # Estadísticas
        running_loss += loss.item() * images.size(0)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
    
    epoch_loss = running_loss / total
    epoch_acc = correct / total
    
    return epoch_loss, epoch_acc
```

## Parte 8: Función de evaluación

```python
# TODO: Implementar la función de evaluación
def evaluate(model, data_loader, criterion, device):
    """
    Evalúa el modelo en un conjunto de datos
    """
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0
    
    with torch.no_grad():  # No calcular gradientes
        for images, labels in data_loader:
            # TODO: Mover datos al dispositivo
            images = images.to(device)
            labels = labels.to(device)
            
            # TODO: Forward pass
            outputs = model(images)
            loss = criterion(outputs, labels)
            
            # Estadísticas
            running_loss += loss.item() * images.size(0)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    
    epoch_loss = running_loss / total
    epoch_acc = correct / total
    
    return epoch_loss, epoch_acc
```

## Parte 9: Loop de entrenamiento completo

```python
# TODO: Implementar el loop de entrenamiento completo
# Guardar métricas para graficar
train_losses = []
train_accs = []
val_losses = []
val_accs = []

print("Iniciando entrenamiento...\n")
best_val_acc = 0.0

for epoch in range(EPOCHS):
    print(f"{'='*60}")
    print(f"Época {epoch+1}/{EPOCHS}")
    print(f"{'='*60}")
    
    # TODO: Entrenar una época
    train_loss, train_acc = train_epoch(
        model, train_loader, criterion, optimizer, DEVICE
    ) # llamar a train_epoch
    
    # TODO: Evaluar en validation
    val_loss, val_acc = evaluate(
        model, val_loader, criterion, DEVICE
    ) # llamar a evaluate con val_loader
    
    # Guardar métricas
    train_losses.append(train_loss)
    train_accs.append(train_acc)
    val_losses.append(val_loss)
    val_accs.append(val_acc)
    
    # Imprimir resultados
    print(f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f}")
    print(f"Val Loss:   {val_loss:.4f} | Val Acc:   {val_acc:.4f}")
    
    # Guardar mejor modelo
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), 'best_model.pth')
        print(f"✓ Mejor modelo guardado (Val Acc: {val_acc:.4f})")
    
    print()

print("\n¡Entrenamiento completado!")
print(f"Mejor accuracy en validación: {best_val_acc:.4f}")
```

## Parte 10: Visualizar curvas de aprendizaje

```python
# TODO: Graficar las curvas de pérdida y accuracy
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Gráfico de pérdida
ax1.plot(train_losses, label='Train', marker='o')
ax1.plot(val_losses, label='Validation', marker='s')
ax1.set_xlabel('Época')
ax1.set_ylabel('Pérdida')
ax1.set_title('Pérdida durante el Entrenamiento')
ax1.legend()
ax1.grid(True)

# Gráfico de accuracy
ax2.plot(train_accs, label='Train', marker='o')
ax2.plot(val_accs, label='Validation', marker='s')
ax2.set_xlabel('Época')
ax2.set_ylabel('Accuracy')
ax2.set_title('Accuracy durante el Entrenamiento')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```

## Parte 11: Evaluar en conjunto de test

```python
# TODO: Cargar el mejor modelo y evaluar en test
model.load_state_dict(torch.load('best_model.pth'))

test_loss, test_acc = evaluate(model, test_loader, criterion, DEVICE)

print(f"{'='*60}")
print(f"RESULTADOS EN TEST")
print(f"{'='*60}")
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")
```

## Parte 12: Matriz de confusión

```python
# TODO: Generar predicciones para crear la matriz de confusión
model.eval()
all_preds = []
all_labels = []

with torch.no_grad():
    for images, labels in test_loader:
        images = images.to(DEVICE)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        
        all_preds.extend(predicted.cpu().numpy())
        all_labels.extend(labels.numpy())

# TODO: Crear y visualizar la matriz de confusión
cm = confusion_matrix(all_labels, all_preds)

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Matriz de Confusión - CNN en Test')
plt.ylabel('Etiqueta Real')
plt.xlabel('Etiqueta Predicha')
plt.show()

# Reporte de clasificación
print("\nReporte de Clasificación:")
print(classification_report(all_labels, all_preds))
```

## Parte 13: Visualizar predicciones correctas e incorrectas

```python
# TODO: Visualizar ejemplos de predicciones correctas e incorrectas
model.eval()

# Obtener un batch de test
dataiter = iter(test_loader)
images, labels = next(dataiter)
images = images.to(DEVICE)

# Hacer predicciones
with torch.no_grad():
    outputs = model(images)
    _, predicted = torch.max(outputs, 1)

# Mover a CPU para visualización
images = images.cpu().numpy()
predicted = predicted.cpu().numpy()
labels = labels.numpy()

# TODO: Encontrar ejemplos correctos e incorrectos
correct_idx = np.where(predicted == labels)[0][:5]
incorrect_idx = np.where(predicted != labels)[0][:5] # encuentra 5 índices donde predicted != labels

# Visualizar
fig, axes = plt.subplots(2, 5, figsize=(15, 6))

# Predicciones correctas
for i, idx in enumerate(correct_idx):
    axes[0, i].imshow(images[idx].reshape(28, 28), cmap='gray')
    axes[0, i].set_title(f"Real: {labels[idx]}\nPred: {predicted[idx]}", color='green')
    axes[0, i].axis('off')

# Predicciones incorrectas
for i, idx in enumerate(incorrect_idx):
    axes[1, i].imshow(images[idx].reshape(28, 28), cmap='gray')
    axes[1, i].set_title(f"Real: {labels[idx]}\nPred: {predicted[idx]}", color='red')
    axes[1, i].axis('off')

axes[0, 0].set_ylabel('Correctas', fontsize=12, fontweight='bold')
axes[1, 0].set_ylabel('Incorrectas', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

## Parte 15: Visualizar mapas de activación

```python
# TODO: Visualizar los mapas de activación para una imagen de ejemplo
# Esto muestra qué características detecta cada filtro en una imagen específica

# Tomar una imagen de ejemplo
example_img = torch.FloatTensor(X_test[0:1]).to(DEVICE)

# Hook para capturar activaciones
activations = {}

def get_activation(name):
    def hook(model, input, output):
        activations[name] = output.detach()
    return hook

# Registrar hooks
model.conv1.register_forward_hook(get_activation('conv1'))
model.conv2.register_forward_hook(get_activation('conv2'))

# Forward pass
model.eval()
with torch.no_grad():
    output = model(example_img)

# Visualizar la imagen original
fig = plt.figure(figsize=(15, 10))

# Imagen original
plt.subplot(3, 1, 1)
plt.imshow(X_test[0].reshape(28, 28), cmap='gray')
plt.title(f'Imagen Original (Dígito: {y_test[0]})')
plt.axis('off')

# Activaciones de conv1 (primeros 16 canales)
conv1_act = activations['conv1'].cpu().numpy()[0]
for i in range(min(16, conv1_act.shape[0])):
    plt.subplot(3, 16, 16 + i + 1)
    plt.imshow(conv1_act[i], cmap='viridis')
    plt.axis('off')
    if i == 0:
        plt.ylabel('Conv1', fontsize=10)

# Activaciones de conv2 (primeros 16 canales)
conv2_act = activations['conv2'].cpu().numpy()[0]
for i in range(min(16, conv2_act.shape[0])):
    plt.subplot(3, 16, 32 + i + 1)
    plt.imshow(conv2_act[i], cmap='viridis')
    plt.axis('off')
    if i == 0:
        plt.ylabel('Conv2', fontsize=10)

plt.tight_layout()
plt.show()
```
