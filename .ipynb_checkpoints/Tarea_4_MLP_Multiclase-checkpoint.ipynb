# Tarea 2: Red Neuronal MLP para Clasificación Multiclase
## Fundamentos de Inteligencia Artificial

**Objetivo**: Implementar y optimizar una red neuronal multicapa (MLP) para clasificar los 10 dígitos de MNIST, explorando diferentes arquitecturas e hiperparámetros.

**Nivel**: Intermedio

**Instrucciones**: Completa las celdas marcadas con `# TODO`

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import time

np.random.seed(42)
```

## Parte 1: Cargar y preparar MNIST completo (10 clases)

```python
print("Cargando MNIST completo...")
mnist = fetch_openml('mnist_784', version=1, parser='auto')
X, y = mnist["data"].values, mnist["target"].values.astype(int)

# Usar un subconjunto para acelerar entrenamiento
n_samples = 20000
indices = np.random.choice(len(X), n_samples, replace=False)
X = X[indices]
y = y[indices]

print(f"Dataset: {X.shape}")
print(f"Distribución de clases:\n{np.bincount(y)}")
```

## Parte 2: Visualización de ejemplos por clase

```python
# TODO: Visualizar un ejemplo de cada dígito (0-9)
fig, axes = plt.subplots(2, 5, figsize=(15, 6))

for digit in range(10):
    # COMPLETA: Encuentra el primer índice donde y == digit
    idx = (y == digit).argmax() 
    
    ax = axes[digit // 5, digit % 5]
    # COMPLETA: Muestra la imagen
    image = X[idx].reshape(28, 28) if X[idx].ndim == 1 else X[idx]
    ax.imshow(image, cmap='gray')
    ax.set_title(f"Dígito: {digit}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```

## Parte 3: Preparación de datos

```python
# TODO: Dividir en train (70%), validation (15%) y test (15%)
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, 
    test_size=0.15,   # 15% para test
    random_state=42, 
    stratify=y
)

X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp,
    test_size=0.176,  # ~15% del total para validation
    random_state=42,
    stratify=y_temp
)

# TODO: Normalizar usando StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

print(f"Train: {X_train.shape[0]}, Val: {X_val.shape[0]}, Test: {X_test.shape[0]}")
```

## Parte 4: Entrenar MLP básico

```python
# TODO: Crear un MLP con arquitectura (128, 64)
# Parámetros: hidden_layer_sizes=(128, 64), max_iter=50, 
#             random_state=42, verbose=True

print("Entrenando MLP básico...")
mlp_basic = MLPClassifier(
    hidden_layer_sizes=(128, 64),
    max_iter=50,
    random_state=42,
    verbose=True
)

start_time = time.time()
mlp_basic.fit(X_train_scaled, y_train)
train_time = time.time() - start_time

# Evaluar
train_acc = accuracy_score(y_train, mlp_basic.predict(X_train_scaled))
val_acc = accuracy_score(y_val, mlp_basic.predict(X_val_scaled))

print(f"\nTiempo de entrenamiento: {train_time:.2f}s")
print(f"Accuracy Train: {train_acc:.4f}")
print(f"Accuracy Validation: {val_acc:.4f}")
```

## Parte 5: Visualizar curvas de aprendizaje

```python
# TODO: Graficar la curva de pérdida (loss) durante el entrenamiento
# Pista: mlp_basic.loss_curve_ contiene los valores de pérdida por época

plt.figure(figsize=(10, 5))
#valores de pérdida
plt.plot(mlp_basic.loss_curve_)
plt.title('Curva de Aprendizaje - Pérdida por Época')
plt.xlabel('Época')
plt.ylabel('Pérdida')
plt.grid(True)
plt.show()
```

## Parte 6: Experimentar con diferentes arquitecturas

```python
# TODO: Entrenar 3 modelos con diferentes arquitecturas y comparar
# Arquitecturas sugeridas:
# - Modelo 1: (256, 128, 64) - Red profunda
# - Modelo 2: (512,) - Red ancha de una capa
# - Modelo 3: (64, 64, 64) - Red uniforme

architectures = [
    (256, 128, 64),  # Modelo 1: Red profunda
    (512,),          # Modelo 2: Red ancha de una capa
    (64, 64, 64)     # Modelo 3: Red uniforme
]

results = []

for i, arch in enumerate(architectures, 1):
    print(f"\n{'='*50}")
    print(f"Entrenando Modelo {i}: {arch}")
    print(f"{'='*50}")
    
    # COMPLETA: Crear y entrenar el modelo
    mlp = MLPClassifier(
        hidden_layer_sizes=arch,
        max_iter=50,
        random_state=42,
        verbose=False
    )
    
    start = time.time()
    mlp.fit(X_train_scaled, y_train)
    time_elapsed = time.time() - start
    
    # Evaluar
    train_acc = accuracy_score(y_train, mlp.predict(X_train_scaled))
    val_acc = accuracy_score(y_val, mlp.predict(X_val_scaled))
    
    results.append({
        'architecture': arch,
        'train_acc': train_acc,
        'val_acc': val_acc,
        'time': time_elapsed,
        'model': mlp
    })
    
    print(f"Train Accuracy: {train_acc:.4f}")
    print(f"Val Accuracy: {val_acc:.4f}")
    print(f"Tiempo: {time_elapsed:.2f}s")
```

## Parte 7: Comparar resultados

```python
# TODO: Crear una visualización comparativa de los 3 modelos
# Graficar Train Accuracy vs Validation Accuracy para cada arquitectura

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Gráfico 1: Accuracy
architectures_str = [str(r['architecture']) for r in results]
train_accs = [r['train_acc'] for r in results]
val_accs = [r['val_acc'] for r in results]

x = np.arange(len(architectures_str))
width = 0.35

# COMPLETA: Crear barras para train y validation accuracy
ax1.bar(x - width/2, train_accs, width, label='Train', color='skyblue')
ax1.bar(x + width/2, val_accs, width, label='Validation', color='salmon')
ax1.set_xlabel('Arquitectura')
ax1.set_ylabel('Accuracy')
ax1.set_title('Comparación de Accuracy por Arquitectura')
ax1.set_xticks(x)
ax1.set_xticklabels(architectures_str, rotation=45)
ax1.legend(['Train', 'Validation'])
ax1.grid(True, alpha=0.3)

# Gráfico 2: Tiempo de entrenamiento
times = [r['time'] for r in results]
ax2.bar(x, times, color='green', alpha=0.7)
ax2.set_xlabel('Arquitectura')
ax2.set_ylabel('Tiempo (segundos)')
ax2.set_title('Tiempo de Entrenamiento por Arquitectura')
ax2.set_xticks(x)
ax2.set_xticklabels(architectures_str, rotation=45)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Parte 8: Seleccionar el mejor modelo y evaluar en test

```python
# TODO: Seleccionar el modelo con mejor accuracy en validación
best_model_idx = np.argmax(val_accs) #índice del mejor modelo basado en val_acc
best_model = results[best_model_idx]['model']
best_arch = results[best_model_idx]['architecture']

print(f"Mejor arquitectura: {best_arch}")
print(f"Val Accuracy: {results[best_model_idx]['val_acc']:.4f}")

# Evaluar en test
y_pred_test = best_model.predict(X_test_scaled)
test_acc = accuracy_score(y_test, y_pred_test)

print(f"\n{'='*50}")
print(f"RESULTADO FINAL EN TEST")
print(f"{'='*50}")
print(f"Test Accuracy: {test_acc:.4f}")
```

## Parte 9: Matriz de confusión y análisis de errores

```python
# TODO: Crear y visualizar la matriz de confusión
cm = confusion_matrix(y_test, y_pred_test)

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Matriz de Confusión - Mejor Modelo')
plt.ylabel('Etiqueta Real')
plt.xlabel('Etiqueta Predicha')
plt.show()

# Reporte de clasificación
print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred_test))
```

## Parte 10: Visualizar ejemplos mal clasificados

```python
# TODO: Encontrar y visualizar 10 ejemplos mal clasificados
# Pista: Encuentra índices donde y_test != y_pred_test

misclassified_idx = np.where(y_test != y_pred_test)[0] #encuentra índices mal clasificados

# Seleccionar 10 ejemplos aleatorios de los mal clasificados
if len(misclassified_idx) > 10:
    sample_idx = np.random.choice(misclassified_idx, 10, replace=False)
else:
    sample_idx = misclassified_idx

fig, axes = plt.subplots(2, 5, figsize=(15, 6))

for i, ax in enumerate(axes.flat):
    if i < len(sample_idx):
        idx = sample_idx[i]
        image = X_test[idx].reshape(28, 28) if X_test[idx].ndim == 1 else X_test[idx]
        ax.imshow(image, cmap='gray')
        ax.set_title(f"Real: {y_test[idx]}, Pred: {y_pred_test[idx]}")
        ax.axis('off')

plt.suptitle('Ejemplos Mal Clasificados', fontsize=16)
plt.tight_layout()
plt.show()
```
